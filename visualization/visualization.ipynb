```python
import pandas as pd
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
```

```python
SP_500_DATA_PKL = os.path.join(os.path.dirname(os.getcwd()), 'data/stocks/SP500_data.pkl')
RUSSELL_3000_DATA_PKL = os.path.join(os.path.dirname(os.getcwd()), 'data/stocks/Russell3000_data.pkl')
```

```python
def scatterPlot(train_points, test_points, title):
    pca = PCA(n_components=2)
    pca.fit(np.vstack(train_points))
    train_xy = pca.transform(np.vstack(train_points))
    test_xy = pca.transform(np.vstack(test_points))
    
    
    plt.scatter(train_xy[:,0], train_xy[:,1], c = 'r', label="Training Points")
    plt.scatter(test_xy[:,0], test_xy[:,1], c = 'b', label="Test Points")

    plt.legend(loc="upper left")
    plt.title(title)
    plt.legend()
    plt.show
```

```python
with open(SP_500_DATA_PKL, 'rb') as file:
    symbols, dimensions, series = pickle.load(file)
```

```python
train_points = series[:round(len(series) * .8)]
test_points = series[round(len(series) * .8):]
scatterPlot(train_points, test_points, "S&P500 - PCA Dimension Reduced")
```

```python
with open(RUSSELL_3000_DATA_PKL, 'rb') as file:
    symbols, dimensions, series = pickle.load(file)
```

```python
train_points = series[:round(len(series) * .8)]
test_points = series[round(len(series) * .8):]
scatterPlot(train_points, test_points, "Russell 3000 - PCA Dimension Reduced")
```
